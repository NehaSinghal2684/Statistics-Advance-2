**Q1.** What is hypothesis testing in statistics ?  
**Ans:**Hypothesis testing is a statistical method that uses data from a sample to draw conclusions about a population parameter or a population probability distribution. It is used to make a statistical decision using experimental data. Hypothesis testing involves evaluating two mutually exclusive statements about a population to determine which statement is best supported by the sample data. A statistical hypothesis test typically involves a calculation of a test statistic.

**Q2.**  What is the null hypothesis, and how does it differ from the alternative hypothesis ?  
**Ans:  
**Null Hypothesis (H₀)** :The null hypothesis is a statement that there is no effect, no difference, or no relationship between variables. It assumes that any observed difference is due to sampling or experimental error.

**Alternative Hypothesis (H₁ or Ha)** : The alternative hypothesis is a statement that there is an effect, a difference, or a relationship between variables. It is what you aim to support or prove through your research.

**Key Differences**:  
1) Purpose: The null hypothesis represents the default or status quo, whereas the alternative hypothesis represents a new theory or discovery that researchers want to test.

2) Testing: Statistical tests are conducted to determine whether the observed data provides enough evidence to reject the null hypothesis in favor of the alternative hypothesis.

3) Outcome: If the null hypothesis is rejected, the alternative hypothesis is accepted. If the null hypothesis is not rejected, it means there isn't enough evidence to support the alternative hypothesis.

**Q3.**What is the significance level in hypothesis testing, and why is it important?  
**Ans**:The significance level, often denoted by alpha (α), is a critical concept in hypothesis testing. It represents the threshold at which we determine whether to reject the null hypothesis. Essentially, it defines the probability of making a Type I error, which occurs when we incorrectly reject a true null hypothesis.

It is important because:

1) Decision-Making: It helps researchers decide whether their results are statistically significant. A common significance level is 0.05, which means there's a 5% risk of rejecting the null hypothesis when it's actually true.

2) Controlling Errors: By setting a significance level, researchers control the likelihood of making a Type I error. Lowering the significance level reduces the chance of false positives but increases the chance of Type II errors (failing to reject a false null hypothesis).

3) Comparability: Standard significance levels (like 0.05, 0.01) allow for consistency and comparability across studies, enabling researchers to compare results and replicate findings.

**Q4.**What does a P-value represent in hypothesis testing?   
**Ans:**In hypothesis testing, a P-value (or probability value) is a measure of the strength of the evidence against the null hypothesis. Here's a more detailed breakdown:  
Null Hypothesis (H0): The default assumption that there is no effect or no difference.

Alternative Hypothesis (H1): The assumption that there is an effect or a difference.

The P-value helps you determine the significance of your results in the context of the null hypothesis.

Low P-value (≤ 0.05): Indicates strong evidence against the null hypothesis, so you reject the null hypothesis.

High P-value (> 0.05): Indicates weak evidence against the null hypothesis, so you fail to reject the null hypothesis.

The P-value doesn't tell you the probability that either hypothesis is true, nor does it measure the size of an effect or the importance of a result. Instead, it's a tool to help you make a decision about the null hypothesis.

**Q5.** How do you interpret the P-value in hypothesis testing ?  
**Ans:**In hypothesis testing, a P-value (or probability value) represents the probability of obtaining test results at least as extreme as the ones observed during the study, assuming that the null hypothesis is true. In simpler terms, it's a measure that helps you determine the strength of the evidence against the null hypothesis.

Low P-value (typically ≤ 0.05): This suggests that the observed data is highly unlikely under the null hypothesis. Thus, you may reject the null hypothesis, indicating that there is a significant effect or difference.

High P-value (> 0.05): This suggests that the observed data is not unusual under the null hypothesis. Thus, you fail to reject the null hypothesis, indicating that there is not enough evidence to suggest a significant effect or difference.

**Q6.**What are Type 1 and Type 2 errors in hypothesis testing ?   
**Ans:**In hypothesis testing, Type 1 and Type 2 errors are the two types of errors that can occur when making decisions based on sample data.

Type 1 Error (False Positive): This occurs when we reject the null hypothesis when it is actually true. It's like a false alarm. For example, if a medical test indicates that a person has a disease when they actually do not, this is a Type 1 error. The probability of making a Type 1 error is denoted by α (alpha), which is also known as the significance level.

Type 2 Error (False Negative): This occurs when we fail to reject the null hypothesis when it is actually false. It's like a missed alarm. For example, if a medical test indicates that a person does not have a disease when they actually do, this is a Type 2 error. The probability of making a Type 2 error is denoted by β (beta).

**Q20.**What is an ANOVA test, and what are its assumptions?  
**Ans:**An ANOVA (Analysis of Variance) test is a statistical method used to determine whether there are any statistically significant differences between the means of three or more independent groups. It's useful for comparing more than two groups to see if at least one group mean is different from the others.

The key assumptions of ANOVA are:

1)Independence: The samples from different groups must be independent of each other. That means the data collected in one group should not influence the data collected in another group.

2)Normality: The data within each group should be approximately normally distributed. This is important for the validity of the test results.

3)Homogeneity of Variances: The variances within each group should be roughly equal. This assumption is also known as homoscedasticity.

**Q21.**What are the different types of ANOVA tests?  
**Ans:**ANOVA, or Analysis of Variance, is a collection of statistical techniques used to compare means among three or more groups. Here are some of the different types of ANOVA tests:

1)One-Way ANOVA: This test compares the means of three or more unrelated groups based on one independent variable. For example, comparing test scores among different teaching methods.

2)Two-Way ANOVA: This test examines the influence of two different categorical independent variables on one continuous dependent variable. It also explores the interaction between the two independent variables. For example, evaluating the effect of diet and exercise on weight loss.

3)Repeated Measures ANOVA: This test is used when the same subjects are measured multiple times under different conditions. It's helpful for analyzing data where the dependent variable is measured repeatedly over time. For example, assessing the impact of a drug on blood pressure at different time points.

4)Mixed-Design ANOVA: This test combines features of both between-subjects and within-subjects designs. It’s used when you have both independent groups and repeated measures. For example, studying the effect of a new teaching method on test scores while considering the factor of time (pre-test and post-test).

**Q1.**Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results ?  
**Ans:**

```python
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, significance_level=0.05):
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    z_score = (sample_mean - population_mean) / standard_error
    p_value = stats.norm.sf(abs(z_score)) * 2  # Two-tailed test

    print(f'Sample Mean: {sample_mean}')
    print(f'Population Mean: {population_mean}')
    print(f'Standard Error: {standard_error}')
    print(f'Z-Score: {z_score}')
    print(f'P-Value: {p_value}')
    if p_value < significance_level:
        print(f'Reject the null hypothesis (p-value: {p_value})')
    else:
        print(f'Fail to reject the null hypothesis (p-value: {p_value})')
sample_data = [12, 14, 16, 15, 15, 13, 16, 14, 13, 15]
population_mean = 14
population_std = 2
z_test(sample_data, population_mean, population_std)
```

**Q3.**Implement a one-sample Z-test using Python to compare the sample mean with the population mean?   
**Ans:**

```python
import numpy as np
import scipy.stats as stats
sample = np.array([10, 12, 9, 11, 13, 9, 10, 12, 11, 10])
population_mean = 10.5
sample_mean = np.mean(sample)

sample_std = np.std(sample, ddof=1)
n = len(sample)
standard_error = sample_std / np.sqrt(n)
z_score = (sample_mean - population_mean) / standard_error
p_value = stats.norm.sf(abs(z_score)) * 2

print(f"Sample Mean: {sample_mean}")
print(f"Population Mean: {population_mean}")
print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")
```

**Q4.** Perform a two-tailed Z-test using Python and visualize the decision region on a plot .  
**Ans:**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
null_mean = 50
sample_mean = 55
pop_std = 10
sample_size = 30
z_score = (sample_mean - null_mean) / (pop_std / np.sqrt(sample_size))
alpha = 0.05
critical_value = stats.norm.ppf(1 - alpha / 2)
print(f"Z-score: {z_score}")
print(f"Critical value: ±{critical_value}")
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)
plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x < -critical_value) | (x > critical_value), color='grey', alpha=0.5, label='Rejection Region')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')
plt.title('Two-tailed Z-test Decision Region')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

**Q5.**Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing .  
**Ans:**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def hypothesis_testing(sample_mean, sample_std, population_mean, alpha, n):

    z_alpha = stats.norm.ppf(1 - alpha/2)

    standard_error = sample_std / np.sqrt(n)

    ci_lower = sample_mean - z_alpha * standard_error
    ci_upper = sample_mean + z_alpha * standard_error

    type_1_error = alpha

    z_beta = (population_mean - sample_mean) / standard_error
    beta = stats.norm.cdf(z_beta + z_alpha) - stats.norm.cdf(z_beta - z_alpha)
    x = np.linspace(sample_mean - 4 * standard_error, sample_mean + 4 * standard_error, 1000)
    y = stats.norm.pdf(x, sample_mean, standard_error)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='Sample Distribution')

    plt.axvline(ci_lower, color='r', linestyle='dashed', linewidth=1, label='Critical Value (Lower)')
    plt.axvline(ci_upper, color='r', linestyle='dashed', linewidth=1, label='Critical Value (Upper)')

    plt.fill_between(x, y, where=(x < ci_lower) | (x > ci_upper), color='red', alpha=0.2, label='Type 1 Error Area')
    plt.fill_between(x, y, where=(x >= population_mean - z_alpha * standard_error) & (x <= population_mean + z_alpha * standard_error), color='blue', alpha=0.2, label='Type 2 Error Area')

    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.legend()
    plt.show()

    return type_1_error, beta
sample_mean = 5.5
sample_std = 1.2
population_mean = 5.0
alpha = 0.05
n = 30
type_1_error, type_2_error = hypothesis_testing(sample_mean, sample_std, population_mean, alpha, n)
print(f'Type 1 Error (alpha): {type_1_error}')
print(f'Type 2 Error (beta): {type_2_error}')
```

**Q6.**Write a Python program to perform an independent T-test and interpret the results.  
**Ans:**

```python
import numpy as np
from scipy import stats

# Sample data
group1 = [12, 14, 16, 18, 19, 22, 24, 25, 27, 30]
group2 = [11, 15, 17, 19, 20, 23, 26, 28, 29, 31]
t_statistic, p_value = stats.ttest_ind(group1, group2)
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("We reject the null hypothesis. There is a significant difference between the two groups.")
else:
    print("We fail to reject the null hypothesis. There is no significant difference between the two groups.")
```

**Q7.**Perform a paired sample T-test using Python and visualize the comparison results .

**Q9.**Write a Python function to calculate the confidence interval for a sample mean and explain its significance.  
**Ans:**

```python
import numpy as np
import scipy.stats as st

def confidence_interval(data, confidence=0.95):
    data = np.array(data)
    mean = np.mean(data)
    sem = st.sem(data)
    margin_of_error = sem * st.t.ppf((1 + confidence) / 2., len(data) - 1)
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return lower_bound, upper_bound
```

**Significance of Confidence Intervals :**  
A confidence interval (CI) for a sample mean provides a range of values within which we can be confident that the true population mean lies. The confidence level (e.g., 95%) indicates the probability that the interval contains the true mean. Here are a few key points:

1)Precision: A narrower confidence interval suggests more precise estimates of the population mean.

2)Uncertainty: Wider intervals indicate greater uncertainty about the true mean.

3)Sample Size: Larger sample sizes generally result in narrower confidence intervals, reflecting more reliable estimates.

4)Application: Confidence intervals are widely used in scientific research, business decision-making, and quality control to provide insights into the reliability of sample estimates.

**Q10.**Write a Python program to calculate the margin of error for a given confidence level using sample data.  
**Ans:**

```python
import math
def margin_of_error(sample_std_dev, sample_size, confidence_level):
    # Z-scores for common confidence levels
    z_scores = {
        0.9: 1.645,
        0.95: 1.96,
        0.99: 2.576
    }
    if confidence_level not in z_scores:
        raise ValueError("Unsupported confidence level. Use 0.9, 0.95, or 0.99.")
    standard_error = sample_std_dev / math.sqrt(sample_size)
    margin = z_scores[confidence_level] * standard_error

    return margin
sample_std_dev = 15.0
sample_size = 100
confidence_level = 0.95
margin = margin_of_error(sample_std_dev, sample_size, confidence_level)
print(f"Margin of Error: {margin:.2f}")
```

**Q13.**Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.  
**Ans:**

```python
import numpy as np

def calculate_expected_frequencies(observed):
    observed = np.array(observed)
    row_totals = observed.sum(axis=1)
    col_totals = observed.sum(axis=0)
    grand_total = observed.sum()
    expected = np.outer(row_totals, col_totals) / grand_total

    return expected
observed_data = [
    [10, 20, 30],
    [20, 30, 50],
    [30, 50, 80]
]
expected_frequencies = calculate_expected_frequencies(observed_data)
print("Observed Data:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_frequencies)
```

**Q16.**Implement an F-test using Python to compare the variances of two random samples.  
**Ans:**

```python
import numpy as np
from scipy.stats import f

np.random.seed(0)
sample1 = np.random.normal(loc=20, scale=5, size=100)
sample2 = np.random.normal(loc=22, scale=7, size=100)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

F = var1 / var2

# Calculate the degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1
p_value = 1 - f.cdf(F, df1, df2)

print("F-statistic:", F)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

**Q19.**Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.  
**Ans:**

```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

def check_anova_assumptions(data, group_column, value_column):
    groups = data[group_column].unique()
    group_data = [data[data[group_column] == group][value_column] for group in groups]

    def check_normality(group):
        k2, p = stats.normaltest(group)
        return p > 0.05
    normality_results = [check_normality(group) for group in group_data]
    def check_independence(data):
        model = sm.OLS(data[value_column], sm.add_constant(data[group_column])).fit()
        residuals = model.resid
        plt.figure()
        plt.plot(residuals)
        plt.title('Residuals Plot for Independence Check')
        plt.xlabel('Observation')
        plt.ylabel('Residuals')
        plt.show()
        dw = sm.stats.stattools.durbin_watson(residuals)
        return 1.5 < dw < 2.5
    independence_result = check_independence(data)
    def check_equal_variance(group_data):
        stat, p = stats.levene(*group_data)
        return p > 0.05
    equal_variance_result = check_equal_variance(group_data)
    results = {
        "Normality": normality_results,
        "Independence": independence_result,
        "Equal Variance": equal_variance_result
    }
    return results
```

**Q24.**Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.  
**Ans:**

```python
import numpy as np
from scipy.stats import chi2

data = [10, 12, 9, 15, 8, 11, 10, 14, 13]
n = len(data)
s2 = np.var(data, ddof=1)
sigma0_squared = 10
chi_square_value = (n - 1) * s2 / sigma0_squared
alpha = 0.05
df = n - 1
chi_square_critical_lower = chi2.ppf(alpha / 2, df)
chi_square_critical_upper = chi2.ppf(1 - alpha / 2, df)
p_value = 2 * min(chi2.cdf(chi_square_value, df), 1 - chi2.cdf(chi_square_value, df))
print(f"Chi-square value: {chi_square_value:.2f}")
print(f"Chi-square critical values: {chi_square_critical_lower:.2f}, {chi_square_critical_upper:.2f}")
print(f"p-value: {p_value:.4f}")
if chi_square_value < chi_square_critical_lower or chi_square_value > chi_square_critical_upper:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

**Q25.**Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.  
**Ans:**

```python
import numpy as np
import scipy.stats as stats

def z_test_proportion(success_a, size_a, success_b, size_b):
    # Proportion of successes
    p1 = success_a / size_a
    p2 = success_b / size_b
    p_pool = (success_a + success_b) / (size_a + size_b)

    se = np.sqrt(p_pool * (1 - p_pool) * (1/size_a + 1/size_b))

    z = (p1 - p2) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    return z, p_value
success_a = 60
size_a = 100
success_b = 50
size_b = 100
z, p_value = z_test_proportion(success_a, size_a, success_b, size_b)
print(f'Z-score: {z}')
print(f'P-value: {p_value}')
```
